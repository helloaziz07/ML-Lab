```python
# Importing Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset

df = pd.read_csv("Customers.csv")
```

```python
# Printing first 5 rows

df.head()
```

```python
# printing the whole dataset

df
```

```python
# PLOTTING : SCATTER PLOT
```

```python
plt.figure(figsize=(8,9))
plt.scatter(x=df['Annual Income ($)'], y=df['Spending Score (1-100)'], color="indigo")

plt.xlabel('Annual Income ($)', fontsize=12, fontweight='bold')
plt.ylabel('Spending Score (1-100)', fontsize=12, fontweight='bold')
plt.title('Customer Segmentation Scatter Plot', fontsize=15, fontweight='bold')

plt.show()
```

```python
# PLOTTING : BAR PLOT
```

```python
plt.figure(figsize=(20,6))
sns.barplot(x=df['Age'], y=df['Annual Income ($)'], color="crimson")
plt.xticks(rotation=90)

plt.xlabel('Age', fontsize=12, fontweight='bold')
plt.ylabel('Annual Income ($)', fontsize=12, fontweight='bold')
plt.title('Customer Segmentation - Annual Income by Age', fontsize=15, fontweight='bold')
plt.show()
```

```python
# PLOTTING : BOX PLOT
```

```python
sns.boxplot(x=df['Profession'], y=df['Spending Score (1-100)'], color="teal")
plt.xticks(rotation=45)

plt.xlabel('Profession', fontsize=12, fontweight='bold')
plt.ylabel('Spending Score (1-100)', fontsize=12, fontweight='bold')
plt.title('Customer Segmentation - Spending Score by Profession', fontsize=15, fontweight='bold')
plt.show()
```

```python
# PLOTTING : PIE CHART
```

```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(8, 8))
gender_counts.plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=90)
plt.title('Gender', fontsize=15, fontweight='bold')
plt.ylabel('')
plt.show()
```

```python
# PLOTTING : LINE CHART
```

```python
data_sorted = df.sort_values(by='Gender')

plt.figure(figsize=(10, 6))
plt.plot(data_sorted['Gender'], data_sorted['Age'], marker='o', linestyle='-', color='b', label='Age')

plt.title('Age', fontsize=15, fontweight='bold')
plt.xlabel('Gender', fontsize=12, fontweight='bold')
plt.ylabel('Age', fontsize=12, fontweight='bold')

plt.grid(True)
plt.legend()
plt.show()
```
