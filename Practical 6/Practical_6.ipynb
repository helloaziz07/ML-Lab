```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.25)

print(f"x Training dataset: {x_train.shape}")
print(f"y Training dataset: {y_train.shape}")
print(f"x test dataset: {x_test.shape}")
print(f"y test dataset: {y_test.shape}")
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.coef_ #w
```

```python
model.intercept_ #b
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred.shape
```

```python
# PLOTTING GRAPHS
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_test, y_pred, color='red', linewidth=3)
plt.title("Temperature vs Year")
plt.xlabel("Year")
plt.ylabel("Temperature")
plt.show()
```

```python
sns.regplot(data=df, x=x_train, y=y_train, color='blue', line_kws={"color":"red", "linewidth":2})
```

```python
# Calculating MSE, MAE, R-Square

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(f"MSE:  {mean_squared_error(y_test,y_pred)}")
print(f"MAE:  {mean_absolute_error(y_test,y_pred)}")
print(f"R-Sqaure :  {r2_score(y_test,y_pred)}")
```
